```python
import pandas as pd

df_companies = pd.read_csv("data/df_text_cleaned.csv")
list_of_ciks = list(set(df_companies["cik"]))
```

```python
import requests
import json

dict_of_ticker_cik = {}
for cik in list_of_ciks:
    resp = json.loads(requests.get(f"http://localhost:8000/api/v1/company/{cik}").text)
    if resp["code"] == 200:
        if resp["data"][0]["ticker"] in dict_of_ticker_cik.keys():
            print("problem")
        else:
            dict_of_ticker_cik[resp["data"][0]["ticker"]] = cik
```

## Check overview in Alpha Vantage

```python
from collections import defaultdict

# Fiscal year end can be taken from here
dict_of_sector_industries = defaultdict(lambda: defaultdict(int))
for ticker, cik in dict_of_ticker_cik.items():
    url = f'https://www.alphavantage.co/query?function=OVERVIEW&symbol={ticker}&apikey=DV3U4RGOYLYSQHN8'
    r = requests.get(url)
    data = r.json()
    try:
        print(ticker)
        print(data["Sector"])
        print(data["Industry"])
        dict_of_sector_industries[data["Sector"]][data["Industry"]] += 1
    except Exception as e:
        print(f"Problem for {ticker}")
    print('-------------')
```
