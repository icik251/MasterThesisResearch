## Get and plot stock prices for CIK 2178

```python
import requests
import json

curr_cik = 2178
    
resp = json.loads(requests.get(f"http://localhost:8000/api/v1/stock_price/{curr_cik}").text)
resp_infl = json.loads(requests.get(f"http://localhost:8000/api/v1/stock_price/inflation/{curr_cik}").text)

if resp['code'] == 200:
    stock_price_list = resp['data']
else:
    print(resp['message'])
    
if resp_infl['code'] == 200:
    stock_price_list_infl = resp_infl['data']
else:
    print(resp_infl['message'])
    
```

```python
import plotly.express as px
import pandas as pd

df_adj_close = pd.DataFrame(stock_price_list)
df_adj_inflation = pd.DataFrame(stock_price_list_infl)
df_merged = pd.merge(df_adj_close, df_adj_inflation, how="inner", on="timestamp", suffixes=(None, "_inflation"))
```

```python
fig = px.line(df_merged, x="timestamp", y=["adjusted_close", "adjusted_close_inflation"], title=f"Adjusted close and adjusted inflation for CIK: {curr_cik}")
fig.show()
```
